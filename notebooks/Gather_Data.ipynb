```python
## GET YOBIT DATA!
EX = ccxt.yobit()
coin_list = EX.fetchMarkets()
```

```python
# the unpythonic way!
symbol_list = []
for i in coin_list:
    coin, base = i['symbol'].split('/') 
    if 'BTC' in base:
        symbol_list.append(i['symbol'])
print(len(symbol_list))
```

```python
# print(symbol_list)
# PARSE MARKET DATA
i = 0
target = len(coin_list)
round = 0
high_volume_set = []
everything_else = []
```

```python
# THIS WORKS ... KINDA!
batch = 2
while i <= target:
    try:
        shorter_list = symbol_list[i:i+batch]
    except:
        shorter_list = symbol_list[i:-1]
    i += batch
    round += 1
    ticker = EX.fetchTickers([x for x in shorter_list])
    for coin in ticker:
        price = ticker[coin]['last']
        vol = ticker[coin]['quoteVolume']
        data = {
            'coin': coin,
            'price': price,
            'vol': vol
        }
        if vol > 1:
            high_volume_set.append(data)
        else:
            everything_else.append(data)
    # break
    print('Sleeping through tickers round# {}'.format(round))
    time.sleep(1.35)
```
